```python
import re

def preprocess_sentence(w):
    w = w.lower().strip()
    
    # spacing between a word and the punctuation following it
    # eg: "he is a boy." => "he is a boy ." 
    # Reference:- https://stackoverflow.com/questions/3645931/python-padding-punctuation-with-white-spaces-keeping-punctuation
    w = re.sub(r"([?.!,])", r" \1 ", w) # add space
    w = re.sub(r'[" "]+', " ", w) # remove multiple spaces and quatations
    
    # replacing everything with space except important symbols
    w = re.sub(r"[^a-zA-Zа-яА-я0-9?.!,]+", " ", w).strip()
    
    # adding a start and an end token to the sentence
    # so that the model know when to start and stop predicting.
    w = '_start_ ' + w + ' _end_'
    return w

with open('test.ru.txt', 'r') as f:
    lines = f.readlines()
    lines = '\n'.join([preprocess_sentence(l) for l in lines])
    with open('test.txt', 'w') as fw:
        fw.write(lines)
```

```python
max_len = 100

with open('train-300k.tsv', 'r') as f:
    with open('train-300k-clean.tsv', 'w') as f_o:
        for line in f:
            i, o = line.split('\t')
            li, lo = len(i.split(' ')),  len(o.split(' ')) 
            if 0.4 < li / lo < 3 and li < max_len and lo < max_len:
                f_o.write(line)
```

```python
max_lines = 100000

with open('train-300k-clean.tsv', 'r') as f:
    with open('train-300k-clean-orig.tsv', 'w') as f_o:
        for i, line in enumerate(f):
            li, lo = line.split('\t')
            f_o.write(li[8:-5] + '\t' + lo[8:-6] + '\n')
            if i > max_lines:
                break
```
